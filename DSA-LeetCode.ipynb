```python
##Q.2 Input: l1 = [2,4,3], l2 = [5,6,4]
#Output: [7,0,8]
#Explanation: 342 + 465 = 807.
```

```python
l1=[2,4,3]
l2=[5,6,4]
a="".join([str(i) for i in l1])
a=int(a)
b="".join([str(y) for y in l2])
b=int(b)
ans1=a+b
ans2=str(ans1)
ans=[]
for x in ans2:
    o=int(x)
    ans.append(o)
ans.reverse()
print(ans)
```

```python
x=[1,3,4,6,5]
x1=[2]
l=set(x+x1)
l=list(l)
count=len(l)
if count%2==0:
    pos_1=(count/2)-1
    pos_1=int(pos_1)
    pos_2=((count/2)-1)+1
    pos_2=int(pos_2)
    ans1=(l[pos_1]+l[pos_2])/2
    print(ans1)
else:
    pos_1=(count/2)-0.5
    pos_1=int(pos_1)
    print(l[pos_1])
```

```python
#Q.7 
#x=2
#n=6
```

```python
x=2
n=6
temp=2
l=[]
for i in range(1,n+1):
    if i%2==0:
        temp1=-x**temp
        l.append(temp1)
        temp+=2
    else:
        temp1=x**temp
        l.append(temp1)
        temp+=2
ans=sum(l)
print(ans)
```
